```python
df = df.rename(columns={'Two Wheelers (Category L1 & L2 as per Central Motor Vehicles Rules': 'L1_L2',
                        'Two Wheelers (Category L2 (CMVR))': 'L2',
                        'Two Wheelers (Max power not exceeding 250 Watts)': 'L2_250W',
                        'Three Wheelers (Category L5 slow speed as per CMVR)': 'L5_slow',
                        'Three Wheelers (Category L5 as per CMVR)': 'L5',
                        'Passenger Cars (Category M1 as per CMVR)': 'M1',
                        'Buses': 'B1',
                        'Total in state':'total'})
```

```python
x=df.groupby("State").size()
plt.figure(figsize=(10,6))
x.plot(kind="pie",autopct="%.2f")
```

```python

def count_for_different_states(c1,c2):
    df_group2=df.groupby([c1,c2]).size()

    x=[]
    y=[]
    for key in df_group2.keys():
        x.append(key[0])
        y.append(key[1])
    plt.bar(x,y,width=0.6,align="edge")
    plt.xticks(x,rotation="vertical")
    plt.xlabel("States")
    plt.ylabel("Count")
    plt.show()
```

```python

for col in df.columns:
    if(col!="State" ):
        
        plt.title("Count of "+col+" among states")
        count_for_different_states("State",col)
```

```python
labels=["L1_L2","L2","L2_250W","L5_slow","L5","M1","B1"]
sizes=[
    df["L1_L2"].sum(),
    df["L2"].sum(),
    df["L2_250W"].sum(),
    df["L5_slow"].sum(),
    df["L5"].sum(),
    df["M1"].sum(),
    df["B1"].sum()
]
colors = ['lightpink', 'yellow', 'blue', 'orange', 'pink', 'violet', 'lightskyblue']
explode = (0, 0, 0.1, 0, 0, 0, 0)
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct='%1.1f%%', shadow=True,startangle=180)
plt.title("Percentage of each vehicle in the country")
plt.show()
```
